```python
# lets look at this data:
df = pd.read_csv("attendance.csv")
df.head()
```

```python
#change the column name
df = df.rename(columns = {"Unnamed: 0" : "name"})
df.head()
```

```python
#melt the df
df = df.melt(id_vars = ['name'], var_name = "day", value_name = "grade")
df.head()
```

```python
#use map to map letter grade with numerica values
df["grade"]= df.grade.map({"P":1,"A":0,"T":0.9,"H":0.5})
df.head()
```

```python
#use group by to group each by names and find the mean of thier grades
df.groupby('name').grade.mean()
```

3. Cake Recipes

Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
Tidy the data as necessary.
Which recipe, on average, is the best? recipe b
Which oven temperature, on average, produces the best results? 275
Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
#load csv
cake_recipes = pd.read_csv("cake_recipes.csv")
cake_recipes
```

```python
#make two columns from recipe:position column
cake_recipes[["recipe", "position"]] = cake_recipes['recipe:position'].str.split(':', expand = True)
cake_recipes.head()
```

```python
#drop the unnesseary column now
cake_recipes = cake_recipes.drop(columns = "recipe:position")
cake_recipes
```

```python
cake_recipies = cake_recipes.melt(id_vars = ['recipe', 'position'], var_name = "oven_temp",
                  value_name="tastiness_score")
                                
cake_recipies.head()                               
```

```python
#3c. Which recipe, on average, is the best? recipe b
cake_recipies.groupby("recipe").tastiness_score.mean()
```

```python
#can use idxmax also
cake_recipies.groupby("recipe").tastiness_score.mean().idxmax()
```

```python
#4d. Which oven temperature, on average, produces the best results? 
cake_recipies.groupby("oven_temp").tastiness_score.mean().idxmax()
```

```python
#4e Which combination of recipe, rack position, and temperature gives the best result?
cake_recipies[cake_recipies.tastiness_score == cake_recipies.tastiness_score.max()]
```
